### Bài 1: Tìm số lớn nhất của 1 list

```python
lst = list(map(int, input().split()))
max = lst[0]

for i in lst:
    if max < i:
        max = i
print(max)
```

### Bài 3: Tìm giá trị trung bình của các số nguyên dương trong list

```python
lst = list(map(int, input().split()))
s = 0
count = 0

for i in lst:
    if i > 0:
        s += i
        count += 1

if count > 0:
    print(round((s / count), 2))

```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
lst = list(map(int, input().split()))
if lst[0:len(lst)] == lst[::-1]:
    print("YES")
else:
    print("NO")
```
