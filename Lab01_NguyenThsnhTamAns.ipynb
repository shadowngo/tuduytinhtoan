### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
n = int(input())
a = 0

for k in range(1, n):
    if k > (n - 1) / 2:
        a += 1
print(a)
```

### Problem 8

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
x = input()

if len(x) != 1:
    raise ValueError("Input must be a single character")
elif 'A' <= x <= 'Z':
    print(x.lower())
else:
    print("Input must be an uppercase letter or in A...Z")
```
