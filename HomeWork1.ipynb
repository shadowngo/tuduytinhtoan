```python
from math import pi as M_Pi

a = float(input())
b = float(input())

S = a * b
r = min(a, b) / 2

S_tron = M_Pi * r **2
S_con = S - S_tron

print(f'{S_con:.2f}')

```

```python
a, b, c = map(float, input().split())

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f'{S:.2f}')
else:
    print("Không phải là 3 cạnh của một tam giác.")
```
