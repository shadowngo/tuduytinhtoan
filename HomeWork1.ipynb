```python
# phan dien tich de trong cay

from math import pi as M_Pi

a = float(input())
b = float(input())

S = a * b
r = min(a, b) / 2

S_tron = M_Pi * r **2
S_con = S - S_tron

print(f'{S_con:.2f}')

```

```python
# # Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
# nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
# không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

a, b, c = map(float, input().split())

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f'{S:.2f}')
else:
    print("Không phải là 3 cạnh của một tam giác.")
```

```python
#  Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
# tính tiền điện tháng này cho hộ biết:

n = input()
a, b = map(float, input().split())
c = max(a, b) - min(a, b)
bill = 0

if c > 400:
    bill += c * 3460
elif c > 300:
    bill += c * 3350
elif c > 200:
    bill += c * 2998
elif c > 100:
    bill += c * 2380
elif c > 50:
    bill += c * 2050
else:
    bill += c * 1984

total_bill = bill * 1.08

print(f"Ho va ten: " + n)
print(f'Chi so thang truoc: {min(a, b)}')
print(f'Chi so thang nay : {max(a, b)}')
print(f'Tien phai tra la: {total_bill}')
```
