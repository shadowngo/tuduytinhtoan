Bai 1: Nhap vao hai canh cua mot hinh chu nhat. In ra man hinh dien tich va chu vi cua no.

```python
a, b = map(float, input().split())

def tinh_hinh_chu_nhat(a, b):
    area = a * b
    perimeter = (a + b) * 2
    return area, perimeter

dt, cv = tinh_hinh_chu_nhat(a, b)

print(f'{dt} {cv}')
```

Bai 2: Nhap vao ban kinh hinh tron, tinh dien tich va chu vi.

```python
import math

r = float(input())

def tinh_hinh_tron(r):
    area = math.pi * (r ** 2)
    permeter = 2 * math.pi * r
    return area, permeter

dt, cv = tinh_hinh_tron(r)
print(f'{dt:.2f} {cv:.2f}')
```

Bai 3: Kiem tra tam giac. Tinh dien tich va chu vi neu no la tam giac.

```python
import math

a, b, c = map(float, input().split())
p = (a + b + c) / 2


def tinh_tam_giac(a, b, c):
    area = math.sqrt(p * (p - a) * (p - b) * (p - c))
    permeter = a + b + c
    return area, permeter


A, P = tinh_tam_giac(a, b, c)


def check(a, b, c):
    if a + b <= c or a + c <= b or b + c <= a:
        print('Khong phai la tam giac')
    elif a == b == c:
        print('Day la tam giac deu')
    elif a == b or a == c or b == c:
        print('Day la tam giac can')
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or c**2 + b**2 == a**2:
        print('Day la tam giac vuong')


check(a, b, c)
print(f'A: {A:.2f} and P: {P:.2f} ')
```

Bai 8: Giai hpt tuyen tinh.

```python
a, b, m, c, d, n = map(float, input().split())

D = a * d - b * c
Dx = m * d - n * b
Dy = a * n - c * m

if D == 0 and (Dx or Dy != 0):
    print('He vo nghiem!')
elif D == Dx == Dy == 0:
    print('He co vo so nghiem')
else:
    print('He co nghiem duy nhat la:')
    x = Dx / D
    y = Dy / D
    print(x, y)
```
