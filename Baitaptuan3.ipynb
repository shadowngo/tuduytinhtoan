[Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh

```python
n = input()
a, b = map(float, input().split())
c = max(a, b) - min(a, b)
bill = 0

if c > 400:
    bill += c * 3460
elif c > 300:
    bill += c * 3350
elif c > 200:
    bill += c * 2998
elif c > 100:
    bill += c * 2380
elif c > 50:
    bill += c * 2050
else:
    bill += c * 1984

total_bill = bill * 1.08

print(f"Ho va ten: " + n)
print(f'Chi so thang truoc: {min(a, b)}')
print(f'Chi so thang nay : {max(a, b)}')
print(f'Tien phai tra la: {total_bill}')
```
