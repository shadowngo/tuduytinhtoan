W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return "int"
    except:
        try:
            float(s)
            return "float"
        except:
            return "neither"
        
s = input()
print(detect_number_type(s))
```

W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_poduct_of_others(a, b, c):
    return True if (a == b * c) or (b == a * c) or (c == a * b) else  False

a, b, c = map(int, input().split())
print(is_one_poduct_of_others(a, b, c))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def return_num_in(n):
    count = 0
    for i in list(n):
        if i != 0:
            count += 1
    return count

n = list(map(int, input().split()))
print(return_num_in(n))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).

```python
def giai_thua(n):
    s = 1
    for i in range(1, n + 1):
        s *= i
    return s

def C(n, k):
    if k > n:
        return 0
    return (giai_thua(n) // (giai_thua(k) * giai_thua(n - k)))


n, k = map(int, input().split())
print(C(n, k))
```

W5A31 Viết hàm tính tổng bình phương các số từ 1 đến n

```python
def tong_binh_phuong(n):
    s = 0
    for i in range(1, n + 1):
        s += i**2
    return s

n = int(input())
print(tong_binh_phuong(n))
```
