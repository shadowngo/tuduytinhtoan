W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
import math

n = int(input())

if math.log2(n).is_integer():
    print(f"{n} là số nguyên dương có dạng 2^k")
else:
    print(f"{n} không phải là số nguyên dương có dạng 2^k")
```

```python
n = int(input())

# Powers of 2 must be greater than 0
if n <= 0:
    print('NO')
else:
    # Keep dividing by 2 as long as it's even
    while n % 2 == 0:
        n //= 2
    
    # If we are left with 1, it was a power of 2
    if n == 1:
        print('YES')
    else:
        print('NO')
```

```python
n = int(input())
if n > 0 and (n & (n - 1)) == 0:
    print(f"{n} là số nguyên dương có dạng 2^k")
else:
    print(f"{n} không phải là số nguyên dương có dạng 2^k")
    
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
even = 0
odd = 0

for i in str(n):
    if int(i) % 2 == 0:
        even += 1
    else:        
        odd += 1

print(even, odd)
```

```python
n = int(input())
s = 0
even = 0
odd = 0

while n > 0:
    s = n % 10
    n //= 10
    if s % 2 == 0:
        even += 1
    else:
        odd += 1
print(even, odd)
```

W4A23
Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
s = 0
max = 0

for k in range(1, n + 1):
    s += k
    if s >= n:
        break
    max = k
print(max)
```

```python
n = int(input())
s = 0
i = 0

while True:
    if i < n:
        i += 1
    s += i
    if s >= n:
        break
print(i - 1)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
a = None
b = None

while True:
    n = int(input())
    if n == -1:
        break
    if a == None or a < n:
        a = n
    if b == None or b > n:
        b = n
print(a, b)  
```

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
s = input()
a, b, c = 0, 0, 0

for i in range(len(s)):
    if s[i].islower():
        a += 1
    if s[i].isupper():
        b += 1
    if s[i].isdigit():
        c += 1
print(a, b, c)
```
